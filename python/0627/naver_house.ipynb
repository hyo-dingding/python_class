```python
import pandas as pd
# for page_num in range(1,11):
url = "https://land.naver.com/news/trendReport.naver?page=5"
data = pd.read_html(url)
```

```python
df = data[0]
df.info()
```

```python
temp = df.copy()
```

```python
temp.info()
```

```python
df_temp = temp['제목'].str.replace('%', "")
df_temp
```

```python
regions = ['전국', '서울', '수도권']
for region in regions:
    df_temp = df_temp.str.replace(region, "")
df_temp
```

```python
df_temp = df_temp.str.split(']', expand=True)
df_temp
```

```python
df_temp = df_temp[1].str.split(',', expand=True)
df_temp
```

```python
df_temp.info()
```

```python
df_temp = df_temp.astype(float)
```

```python
df_temp.info()
```

```python
temp[regions] = df_temp
```

```python
temp
```

```python
#df_rate = temp[['등록일','전국','서울','수도권','번호']]
df_rate = temp[['등록일'] + regions + ['번호']]
df_rate.info()
```

```python
df_total_chart = df_rate.copy()
```

```python
df_total_chart
```

```python
df_total_chart['등록일'] = df_total_chart['등록일'].str.replace(".","").astype('datetime64[ms]')
```

```python
df_total_chart
```

```python
# 월별 평균 그룹화 추이 분석
# 년도별 추이 분석
```

```python
df_total_chart['월'] = df_total_chart['등록일'].dt.month
df_total_chart['년'] = df_total_chart['등록일'].dt.year
```

```python
df_total_chart
```

```python
# month_in = int(input('검색할 월입력>>'))
# month_df = df_total_chart.loc[df_total_chart['월']== month_in,]

monthly_avg = df_total_chart.groupby('월')[['전국', '서울', '수도권']].mean().reset_index()
monthly_avg
```

```python
year_avg = df_total_chart.groupby('년')[['전국', '서울', '수도권']].mean().reset_index()
year_avg
```
